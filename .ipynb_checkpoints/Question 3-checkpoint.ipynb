# Question 3.
The dataset Boston from the library MASS consists of 506 median prices of owner-occupoed homes in $1000s (medv) in various places in Boston. Alongside with price, the dataset also provide various geographic and socio-economic information such as:

- crim -- per capita crime rate by town
- zn -- proportion of residential land zoned for lots over 25,000 sq.ft
- indus -- proportion of non-retail business acres per town
- chas -- Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)
- nox -- nitrogen oxides concentration (parts per 10 million)
- rm -- average number of rooms per dwelling
- age -- proportion of owner-occupied units built prior to 1940
- dis -- weighted mean of distances to five Boston employment centres
- rad -- index of accessibility to radial highways
- tax -- full-value property-tax rate per $10,000
- ptratio -- pupil-teacher ratio by town
- black -- 1000(Bkâˆ’0.63)^2 where Bk is the proportion of blacks by town
- lstat -- lower status of the population (percent)

The goal is to find the relations between these factors and the house prices.
1. Analyze the data to get some first impressions and make some preliminary comments.
2. Split randomly (why?) the data into a training and test sets of 80% and 20% of the data respectively. Put a test set meanwhile aside and consider a training set:
    1. Start from the main effects model, verify its adecuacy.
    2. If you're not satisfied, try to add paired interactions, perform transformations if necessary.
    3. Perform model selection w.r.t. various model selection criteria. Compare the resulting models and comment the results.
3. Test and compare the goodness-of-fit of those models on the test set. Comment the results, choose the 'final' model and explain it.

```r
library(MASS)
attach(Boston)
boston = as.data.frame(Boston)
boston$chas = as.factor(boston$chas)
```

## Analyze the data to get some first impressions and make some preliminary comments.

### Correlation matrix

```r
library(corrplot)

corr_matrix<-cor(Boston)
options(repr.plot.width=7, repr.plot.height=7)
corrplot(corr_matrix, type="upper", method = "number")
```

#### From the correlation matrix:

1. Median value of owner-occupied homes (**mdev**)
    - Increases as average number of rooms per dwelling (**rm**) increases
            A house with more rooms is usually more expensive
    - Decreases as percent of lower status population (**lstat**) in the area increases 
            Houses are cheaper when there are more people that 
            cannot afford an expensive house (no demand for expensive huses)
2. Full-value property-tax rate (**tax**)
    - Increases as index of accessibility to radial highways (**rad**) increases
            Access to more highways raises the taxes which make sense because 
            the municipality is investing more money (in higeways) in this area
    - Increases as proportion of non-retail business acres per town (**indus**) increases
            Maybe more businesses needs more maintenence (cleaning, infrastructure, etc..)
            by the municipality, so it needs more money
3. Weighted mean of distances to five Boston employment centres (**dis**)
    - Decreases as proportion of non-retail business acres per town (**indus**) increases
            There are more non-retail business as you get closer to commercial centers
    - Decreases as nitrogen oxides concentration (**nox**) (Air pollution) increases
            There is more nitrogen oxides (air pollution) as you get closer to commercial centers
    - Decreases as proportion of owner-occupied units built prior to 1940 (**age**) increases
            Probably new business owners prefered to open their businesses in areas that
            was already assosiated as commercial centers (because the old businesses)
4. Nitrogen oxides concentration (**nox**) (Air pollution)
    - Increases as proportion of non-retail business acres per town (**indus**)
            There is more nitrogen oxides (air pollution) as you get closer to commercial centers
    - Increases as proportion of owner-occupied units built prior to 1940 (**age**)
            There are more commercial centers around old businneses (which we saw in the previous point)
            so there are more air pollution

## Split randomly (why?) the data into a training and test sets of 80% and 20% of the data respectively. 
- Put a test set meanwhile aside and consider a training set

We split the data randomly because we don't want to train only on certain types of samples and not see at all the other types - overfitting

```r
set.seed(123)
train.indexes <- sample(nrow(boston), nrow(boston) * 0.80)
boston.train <- boston[train.indexes, ]
boston.test <- boston[-train.indexes, ]
```

#### Compare the models we have so far 

```r
crossVal <- function(lm.object)
{
    # Computes the cross-validation (CV), the generalized cross-validation (GCV) and the cross-validation correlation coefficient (R^2_{CV}) for the specified linear model
    res <- lm.object$residuals
    y <- lm.object$fitted.values + lm.object$residuals
    h <- lm.influence(lm.object)$hat
    n <- length(y)
    cv <- mean(res^2/(1 - h)^2)
    gcv <- (n * sum(res^2))/lm.object$df^2
    r2cv <- cor(y, y - res/(1 - h))^2
    return(cbind(cv, gcv, r2cv))
}
```

```r
none <- function(x){ return (x)}

mspe <- function(lm.object, un.trans = none, newdata = boston.test, ...){
    pred.test <- un.trans(predict(lm.object, newdata, ...))
    mspe <- mean((pred.test - boston.test$medv) ^ 2)
    
    return (cbind(mspe))
}
```

```r
lm.maineffects.fit.crossVal <- crossVal(lm.maineffects.fit)
lm.maineffects.xtrans.fit.crossVal <- crossVal(lm.maineffects.xtrans.fit)
lm.pairedint.fit.crossVal <- crossVal(lm.pairedint.fit)
lm.pairedint.xtrans.fit.crossVal <- crossVal(lm.pairedint.xtrans.fit)
lm.pairedint.xytrans.fit.crossVal <- crossVal(lm.pairedint.xytrans.fit)
models <- c("Main effects", "Main effects with x trans", "Paired interactions", "Paired interactions with x trans", "Paired interactions with both, x and y trans")

results <- cbind(models,
                 rbind(lm.maineffects.fit.crossVal, 
                       lm.maineffects.xtrans.fit.crossVal, 
                       lm.pairedint.fit.crossVal, 
                       lm.pairedint.xtrans.fit.crossVal, 
                       lm.pairedint.xytrans.fit.crossVal))
```

```r
lm.maineffects.fit.mspe <- mspe(lm.maineffects.fit)
lm.maineffects.xtrans.fit.mspe <- mspe(lm.maineffects.xtrans.fit)
lm.pairedint.fit.mspe <- mspe(lm.pairedint.fit)
lm.pairedint.xtrans.fit.mspe <- mspe(lm.pairedint.xtrans.fit)

quad <- function(x){ return(x^2)}
lm.pairedint.xytrans.fit.mspe <- mspe(lm.pairedint.xytrans.fit, un.trans = quad)

results <- cbind(results, 
     rbind(lm.maineffects.fit.mspe, 
           lm.maineffects.xtrans.fit.mspe,
           lm.pairedint.fit.mspe,
           lm.pairedint.xtrans.fit.mspe,
           lm.pairedint.xytrans.fit.mspe))
results
```

#### Compare the models we have so far - summary

2. By the r2cv, the last model with the paired interactions and both x & y transformations is a bit better
3. By mspe, the model with the paird interactions and without the transformations is the winner

### Perform model selection w.r.t. various model selection criteria. 
- Compare the resulting models and comment the results.

#### AIC, BIC & RIC

```r
mod.aic <- stepAIC(lm.pairedint.fit,
                   scope=list(upper= medv ~ .^2, lower= ~1), 
                   direction = "both",
                   trace=F,
                   k=2)
# summary(mod.aic)
```

```r
n=nrow(boston.train)

mod.bic <- stepAIC(lm.pairedint.fit,
                   scope=list(upper= medv ~ .^2, lower= ~1), 
                   direction = "both",
                   trace=F,
                   k=log(n))
# summary(mod.bic)
```

```r
p=length(lm.pairedint.fit$coefficients)-1

mod.ric <- stepAIC(lm.pairedint.fit,
                   scope=list(upper= medv ~ .^2, lower= ~1), 
                   direction = "both",
                   trace=F,
                   k=2*log(p))
# summary(mod.ric)
```

```r
# print(anova(lm.pairedint.fit, mod.aic))
# print(anova(lm.pairedint.fit, mod.bic))
# print(anova(lm.pairedint.fit, mod.ric))
```

```r
mod.aic.crossVal <- crossVal(mod.aic)
mod.bic.crossVal <- crossVal(mod.bic)
mod.ric.crossVal <- crossVal(mod.ric)

models <- c("AIC", "BIC", "RIC")

step_results <- cbind(models,
                 rbind(mod.aic.crossVal, 
                       mod.bic.crossVal, 
                       mod.ric.crossVal))

mod.aic.mspe <- mspe(mod.aic)
mod.bic.mspe <- mspe(mod.bic)
mod.ric.mspe <- mspe(mod.ric)

step_results <- cbind(step_results, 
     rbind(mod.aic.mspe, 
           mod.bic.mspe,
           mod.ric.mspe))

rbind(results, step_results)
```

#### Lasso
- starting from the model with all the paired interactions

```r
library(glmnet)

X.lasso <- model.matrix(lm.pairedint.fit)

mod.lasso0.cv <- cv.glmnet(X.lasso,
                           boston.train$medv,alpha=1,
                           nfolds=nrow(X.lasso), #Choose lambda using LOO CV
                           grouped=FALSE)

options(repr.plot.width=5, repr.plot.height=4)
plot(mod.lasso0.cv)
```

```r
mod.lasso <- glmnet(X.lasso, boston.train$medv,alpha=1,lambda=mod.lasso0.cv$lambda.min)
# print(round(predict(mod.lasso,s=mod.lasso0.cv$lambda.min,type="coefficients"),5))
```

```r
#calculate lasso CV
lambda.cv.index = which(mod.lasso0.cv$lambda == mod.lasso0.cv$lambda.min)
mod.lasso.cv = mod.lasso0.cv$cvm[lambda.cv.index][[1]]

#calculate lasso GCV
mod.lasso.rss = sum((predict(mod.lasso, model.matrix(lm.pairedint.fit)) - boston.train$medv)^2)
n = nrow(X.lasso)
p = ncol(X.lasso)
mod.lasso.gcv = (n/(n-p)^2)*mod.lasso.rss

#calculate lasso R2CV
mod.lasso.cv.pred <- lapply(1:nrow(X.lasso), function(i){
    fit <- glmnet(as.matrix(X.lasso[-i,]),
                boston.train$medv[-i],
                alpha = 1,
                lambda = mod.lasso0.cv$lambda.min)
    pred <- predict(fit, t(X.lasso[i,]))[1]

    return(pred)
})
mod.lasso.r2cv = cor(boston.train$medv, as.numeric(mod.lasso.cv.pred))^2

#calculate lasso MSPE
newx <- model.matrix(as.formula(medv ~ .^2), boston.test)
mod.lasso.mspe <- mspe(mod.lasso, newdata = newx)
```

```r
lasso_result = cbind("LASSO", mod.lasso.cv, mod.lasso.gcv, mod.lasso.r2cv, mod.lasso.mspe)
rbind(results, step_results, lasso_result)
```

#### PCR & PLS

```r
pls.crossVal <- function(pls.object, p)
{
    n = nrow(pls.object$fitted.values)
    
    # CV
    cv = pls.object$validation$PRESS[,p]/n

    # GCV
    rss = sum(pls.object$residuals[,,][,p]^2)
    gcv = (n/(n-p)^2)*rss

    # R2CV
    y <- mod.pcr$fitted.values[,,][,p] + mod.pcr$residuals[,,][,p]
    r2cv = cor(y, mod.pcr$validation$pred[,,][,p])^2

    return(cbind(cv, gcv, r2cv))
}
```

```r
library(pls)

mod.pcr = pcr(medv ~ .^2, data=boston.train, scale=T, validation="LOO")
```

```r
summary(mod.pcr)
```

```r
options(repr.plot.width=10, repr.plot.height=5)
par(mfrow = c(1,2))
validationplot(mod.pcr, main="All components")
validationplot(mod.pcr, ncomp = 1:20, main="Zoom into the first 20 components")
```

We can see that after the first 5 components, CV has a negligible improvment. 76.54% of the variance in the data is being explained by these 5 first components.
- Projecting the data on the first 5 eigenvectors will significantly reduce the data dimensions and will explain most of it's variance

```r
# 5 comp PCR CV, GCV, R2CV
pcr_results <- pls.crossVal(mod.pcr, 8)

# 5 comp PCR MSPE
mod.pcr.mspe <- mspe(mod.pcr, ncomp = 8)
```

```r
pcr_results = cbind("PCR - 8 components", pcr_results, mod.pcr.mspe)
rbind(results, step_results, lasso_result, pcr_results)
```
